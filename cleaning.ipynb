```python
import pandas as pd
import numpy as np
import re

def extract_salary_range_min_k(salary):
    # Convert to string to handle both string and float inputs
    salary_str = str(salary) if salary is not None else ''

    # Debug: Print the salary being processed
    print(f"Processing salary: {salary_str}")

    # Return None for undisclosed or unpaid salaries
    if 'disclosed' in salary_str.lower() or 'unpaid' in salary_str.lower():
        return None

    # Handle Crore (Cr) format
    if 'Cr' in salary_str or 'crore' in salary_str.lower():
        match = re.findall(r'(\d+\.?\d*)', salary_str)
        if match:
            return float(match[0]) * 100000  # Convert Cr to k (thousands)

    # Handle Lakh (Lac, Lacs) format
    if 'Lacs' in salary_str or 'Lac' in salary_str or 'lakh' in salary_str.lower():
        match = re.findall(r'(\d+\.?\d*)', salary_str)
        if match:
            return float(match[0]) * 100  # Convert Lacs to k (thousands)

    # Extract numeric value directly if no Cr or Lacs present
    match = re.findall(r'(\d+\.?\d*)', salary_str)
    if match:
        return float(match[0])  # Return numeric in k

    return None  # Return None if no valid salary found

# Apply the function to extract the minimum salary range
df['salary'] = df['salary'].apply(extract_salary_range_min_k)

# Check the extracted salary values
print(df['salary'].head())

# Fill NaN values in the salary column with the mean of the column
mean_salary = df['salary'].mean()
df['salary'] = df['salary'].fillna(mean_salary)
'''
# Define bins and labels for better categorization
salary_bins = [0, 30000, 60000, 90000, 150000, float('inf')]  # Fewer bins for salary ranges
salary_labels = ['Up to 30k', '30k-60k', '60k-90k', '90k-150k', '150k+']

# Bin the salaries into these categories
df['salary_range'] = pd.cut(df['salary'], bins=salary_bins, labels=salary_labels, include_lowest=True)

# Check the unique values in salary_range
print(df['salary_range'].unique())
'''
```

**Location**

```python
# Split the locations by comma and explode into separate rows
df['location'] = df['location'].str.split(',')
df = df.explode('location')

df['location'] = df['location'].str.title().str.strip()  
df['location'] = df['location'].str.replace('Hybrid - ', '', regex=False)  
df['location'] = df['location'].str.replace(r'\s*\(.*?\)\s*', ' ', regex=True)  
df['location'] = df['location'].replace({'Bangalore/Bengaluru': 'Bengaluru', 'Bengaluru ': 'Bengaluru','Bangalore/ Bengaluru':'Bengaluru',
                                         'Bangalore Rural':'Bengaluru'})
df['location'] = df['location'].replace('Delhi / Ncr','NCR')
df['location'] = df['location'].replace('Trivandrum/Thiruvananthapuram','Thiruvananthapuram')
df['location'] = df['location'].replace('Gurgaon/ Gurugram','Gurugram')
df['location'] = df['location'].replace('Gurgaon/Gurugram','Gurugram')
df['location'] = df['location'].replace(['Hyderabad/Secunderabad','Hyderabad/ Secunderabad'],'Hyderabad')
df['location'] = df['location'].replace('Pondicherry/Puducherry','Puducherry')
df['location'] = df['location'].replace(['Nasik', 'Nasik/ Nashik', 'Nasik/Nashik'], 'Nashik')


# Define keywords to drop
keywords = ['India', 'Remote', 'United States', 'Florida', 'Uae', 'Dubai','Saudi Arabia', 'Australia', 'Canada']

# Create a boolean mask and filter the DataFrame
df = df[~df['location'].str.contains('|'.join(keywords), na=False)]

```
